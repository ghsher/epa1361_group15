```python
import pandas as pd
import pareto

INPUT_FILENAME = 'output/20230614T1059_results_seed_ds'

df = pd.read_csv('output/20230614T1059_results_seed_ds.csv')
df
```

```python
file = open(INPUT_FILENAME + '.csv')
tables = [pareto.noattribution(file)]
# 0th row isa header
tables = [pareto.filter_lines(annotatedlines, header=1) for annotatedlines in tables]
tables = [pareto.rowsof(annotatedlines, ',') for annotatedlines in tables]
# we are interested in outcome columns numbered 32-36
tables = [pareto.withobjectives(annotatedrows, range(32, 37)) for annotatedrows in tables]
tagalongs = pareto.eps_sort_solutions(tables, [100, 0.01, 100, 100, 0.01])
```

```python
df_out = pd.DataFrame(columns=list(df.columns))
df_out
```

```python
for row in tagalongs:
    df_out.loc[len(df_out)] = row
```

```python
df_out.to_csv(INPUT_FILENAME + '_pareto.csv')
```
