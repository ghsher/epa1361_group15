```python
from ema_workbench.analysis import parcoords
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("data/results_directed_search_example.csv")
data = data[["A.4 Expected Annual Damage", "A.4 Expected Number of Deaths",
        "Total Infrastructure Costs","Total Expected Annual Damage","Total Expected Number of Deaths"]]
data.head()
```

```python
# get the minimum and maximum values as present in the dataframe
limits = parcoords.get_limits(data)

# we know that the lowest possible value for all objectives is 0
limits.loc[0, ["A.4 Expected Annual Damage", "A.4 Expected Number of Deaths",
               "Total Infrastructure Costs","Total Expected Annual Damage","Total Expected Number of Deaths"]] = 0
# max values
limits.loc[1, ["A.4 Expected Annual Damage"]] = data["A.4 Expected Annual Damage"].max()
limits.loc[1, ["A.4 Expected Number of Deaths"]] = data["A.4 Expected Number of Deaths"].max()
limits.loc[1, ["Total Infrastructure Costs"]] = data["Total Infrastructure Costs"].max()
limits.loc[1, ["Total Expected Annual Damage"]] = data["Total Expected Annual Damage"].max()
limits.loc[1, ["Total Expected Number of Deaths"]] = data["Total Expected Number of Deaths"].max()

paraxes = parcoords.ParallelAxes(limits)
paraxes.plot(data)

plt.show()
```
